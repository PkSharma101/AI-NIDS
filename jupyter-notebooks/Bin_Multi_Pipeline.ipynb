```
def normal(category):
  if category ==1:
    print("Normal Traffic of Cat-{}".format(category))
  else:
    print("Normal Traffic of Cat-{}".format(category))
```

```
def malicious(class_index):
  print("Traffic is Detected as {}".format(classes[class_index][0]))
```
